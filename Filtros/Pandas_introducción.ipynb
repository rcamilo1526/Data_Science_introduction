```python
#borra axis=0:filas any:todas borra columnas con cualquier valor vacio
docentes_sin_nan1= docentes.dropna(axis=0, how='any')
docentes_sin_nan1
```

```python
#borra axis=0:filas all:cualquiera borra filas con todos los datos vacios
docentes_sin_nan2= docentes.dropna(axis=0, how='all')
docentes_sin_nan2
```

```python
#borra axis=1:columnas any:cualquiera borra columnas con algun dato vacio
docentes_sin_nan3= docentes.dropna(axis=1, how='any')
docentes_sin_nan3
```

```python
#llena valores vacios con 3
docentes_nuevo = docentes.fillna(value=3)
docentes_nuevo
```
