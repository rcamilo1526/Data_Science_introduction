```python
import pandas as pd
import numpy as np
pd.options.display.max_rows = 15
```

<b>1.</b> Importar el Data Set “Resultados Saber Pro Competencias Especificas 2018-2” de Datos Abiertos

```python
df=pd.read_csv('Resultados_Saber_Pro_Competencias_Especificas_2018-2.csv')
df
```

<b>2.</b> Imprimir los consecutivos de los estudiantes que en la prueba de GESTIÓN
FINANCIERA obtuvieron un DESEMPEÑO superior a 2

```python
df['ESTU_CONSECUTIVO'][(df['RESULT_NOMBREPRUEBA'] == 'GESTIÓN FINANCIERA') & (df['RESULT_DESEMPENO'] >2)]
```

<b>3.</b> Imprimir los CONSECUTIVOS de los estudiantes que en la prueba de FORMULACIÓN
DE PROYECTOS DE INGENIERÍA obtuvieron un PUNTAJE inferior a 170


```python
df['ESTU_CONSECUTIVO'][(df['RESULT_NOMBREPRUEBA'] == 'FORMULACIÓN DE PROYECTOS DE INGENIERÍA') & (df['RESULT_PUNTAJE'] < 170)]
```

<b>4.</b> Imprima los CONSECUTIVOS de los estudiantes y el NOMBRE DE LA PRUEBA donde
el PUNTAJE sea superior a 200

```python
df[['ESTU_CONSECUTIVO','RESULT_NOMBREPRUEBA']][df['RESULT_PUNTAJE'] > 200]
```

<b>5.</b> Imprima el NOMBRE DE LAS PRUEBAS que obtuvieron DESEMPEÑO superior a 3.
Ojo el nombre de las pruebas no puede estar repetido

```python
#sacando valores unicos y los muestra como una serie en este caso
pd.Series((df['RESULT_NOMBREPRUEBA'][df['RESULT_DESEMPENO'] >3]).unique())
```

<b>6.</b> Utilizando WHERE convierta en NaN todos los registros donde el PUNTAJE no sea
superior a 170 y luego elimine esos registros

```python
df2=df
filtro=df2['RESULT_PUNTAJE']>170
df2=df2.where(filtro,np.nan)
df2=df2.dropna(axis=0, how='all')
df2
```

<b>7.</b> Utilizando MASK convierta en NaN todos los registros donde el DESEMPEÑO no sea
inferior a 3 y luego elimine esos registros.

```python
df3=df
mask=df3['RESULT_DESEMPENO'] >= 3
df3=df3.mask(mask,np.nan)
df3=df3.dropna(axis=0, how='all')
df3
```

<b>8.</b> Utilizando Isin imprima los CONSECUTIVOS de los estudiantes y el NOMBRE
DE LA PRUEBA donde el PUNTAJE sea 160 o 170 o 180 o 190 o 200.

```python
dfisin=df['RESULT_PUNTAJE'].isin([160,170,180,190,200])
df[['ESTU_CONSECUTIVO','RESULT_NOMBREPRUEBA']][dfisin]
```
