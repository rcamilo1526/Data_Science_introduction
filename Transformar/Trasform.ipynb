```python
# EJERCICIO 1
# Crear un ejemplo de pipe con tres funciones.
# La primera y la tercera función deben tener argumentos diferentes al dataframe

def funcion1(dataframe, col,n):
    dataframe[col] = dataframe[col]**n 
    return dataframe
def funcion2(dataframe):
    return dataframe*10
def funcion3(dataframe, col1,col2,nuevacol):
    dataframe[nuevacol] = np.sqrt(dataframe[col1]**2 + dataframe[col2]**2)
    return dataframe
```

```python
print(df.pipe(funcion1,'A',3).pipe(funcion2).pipe(funcion3,'D','E','F'))
```

```python
# EJERCICIO 4

# Genere un dataframe donde se almacene el precio (en pesos colombianos) de algunos productos a través de los meses de un año

meses=['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
dict={'Meses':meses,
      'Detergente':np.random.randint(low=10000, high=18000, size=12),
      'Carne':np.random.randint(low=10000, high=18000, size=12),
      'Carga cafe':np.random.randint(low=500000, high=900000, size=12),
      'Huevo':np.random.randint(low=300, high=800, size=12)}
ex3 = pd.DataFrame(dict)
ex3.set_index('Meses',inplace= True)
ex3
```

```python
# A partir del dataframe anterior genere un nuevo dataframe donde los precios de los productos se encuentren en dolares, euros
# y libras.
def dolar(x):
    return x/3436.65
def euro(x):
    return x/3771.55
def libra(x):
    return x/4226.08
```

```python
# EJERCICIO 5

# Realice el ejercicio 4 pero únicamento convirtiendo de pesos a dolares

print(ex3.transform([dolar,euro,libra]))
```
